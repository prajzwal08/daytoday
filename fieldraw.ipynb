```python
import pandas as pd
```

```python
df_F_1_3_SMST3 = pd.read_csv('/home/khanalp/fieldraw/All-F1_3_SMST3%28z6-25928%29-1747834002/F1_3_SMST3(z6-25928)-Configuration 1-1747834002.3141677.csv')
```

```python

df_F_1_3_SMST3 = df_F_1_3_SMST3.iloc[2:].reset_index(drop=True)
df_F_1_3_SMST3 = df_F_1_3_SMST3.iloc[:, :-4]
```

```python
depths = [2,5,10,20,40,80]
```

```python
new_columns = ['time']
for d in depths:
    new_columns.extend([f'SWC_{d}', f'temp_{d}'])
df_F_1_3_SMST3.columns = new_columns
```

```python
df_F_1_3_SMST3['time'] = pd.to_datetime(df_F_1_3_SMST3['time'])
df_F_1_3_SMST3.set_index('time', inplace=True)
```

```python
df_F_1_3_SMST3
```

```python
df_F_1_4_WPST = pd.read_csv('/home/khanalp/fieldraw/All-F1_4_WPST%28z6-25927%29-1747834077/F1_4_WPST(z6-25927)-Configuration 2-1747834077.9795856.csv')
```

```python

df_F_1_4_WPST = df_F_1_4_WPST.iloc[2:].reset_index(drop=True)
df_F_1_4_WPST = df_F_1_4_WPST.iloc[:, :-4]
```

```python
df_F_1_4_WPST
```

```python
new_columns = ['time']
for d in depths:
    new_columns.extend([f'SWP_{d}', f'temp_{d}'])
df_F_1_4_WPST.columns = new_columns
```

```python
df_F_1_4_WPST['time'] = pd.to_datetime(df_F_1_4_WPST['time'])
df_F_1_4_WPST.set_index('time', inplace=True)
```

```python
df_F_1_3_SMST3 = df_F_1_3_SMST3.apply(pd.to_numeric, errors='coerce')
df_F_1_4_WPST = df_F_1_4_WPST.apply(pd.to_numeric, errors='coerce')
```

```python
import numpy as np

df_F_1_3_SMST3.replace(-99999.0, np.nan, inplace=True)
df_F_1_4_WPST.replace(-99999.0, np.nan, inplace=True)
```

```python
# Find the overlapping time range
start_time = max(df_F_1_3_SMST3.index.min(), df_F_1_4_WPST.index.min())
end_time = min(df_F_1_3_SMST3.index.max(), df_F_1_4_WPST.index.max())

```

```python
df_F_1_3_SMST3 = df_F_1_3_SMST3.sort_index().loc[start_time:end_time]
# Align both dataframes to keep only rows with matching indices
common_index = df_F_1_3_SMST3.index.intersection(df_F_1_4_WPST.index)
df_F_1_3_SMST3 = df_F_1_3_SMST3.loc[common_index]
df_F_1_4_WPST = df_F_1_4_WPST.loc[common_index]
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax1 = plt.subplots(figsize=(12, 5))

ax1.scatter(df_F_1_4_WPST.index, df_F_1_4_WPST['SWP_2'], color='tab:blue', label='SWP_2 (WPST)', s=5)
ax1.set_ylabel('SWP_2', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()
ax2.scatter(df_F_1_3_SMST3.index, df_F_1_3_SMST3['SWC_2'], color='tab:orange', label='SWC_2 (SMST3)', s=5)
ax2.set_ylabel('SWC_2', color='tab:orange')
ax2.tick_params(axis='y', labelcolor='tab:orange')

fig.tight_layout()
plt.title('SWP_2 (WPST) and SWC_2 (SMST3) over Time (Scatter Time Series)')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(df_F_1_3_SMST3['SWC_2'], np.log(-df_F_1_4_WPST['SWP_2']), s=5, alpha=0.5)
ax.set_xlabel('SWC_2 (SMST3)')
ax.set_ylabel('SWP_2 (WPST)')
ax.set_title('SWC_2 vs SWP_2 Scatter Plot')
plt.tight_layout()
plt.show()
```
